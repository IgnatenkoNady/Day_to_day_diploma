# ТЕЛЕКОМ. Предсказание оттока клиентов

## Описание проекта

Оператор связи «ТелеДом» хочет уменьшить отток клиентов. Для этого сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Для определения таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора предоставила персональные данные о некоторых клиентах, информацию об их тарифах и услугах. 

## Задача
Подобрать модель и обучить на этих данных для прогноза оттока клиентов.


## Описание услуг
Оператор предоставляет два основных типа услуг:
1. Стационарнаятелефонная связь. Телефон можно подключить к нескольким линиям одновременно.
2. Интернет. Подключение бывает двух типов: через телефонную линию DSL (англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (англ. fiber optic).

Также абонентам доступен ряд услуг:
1) Интернет-безопасность: антивирус (Device Protection) и блокировка опасных сайтов (Online Security);
2) Выделенная линия технической поддержки (Tech Support);
3) Облачное хранилище файлов для резервного копирования данных (Online Backup);
4) Стриминговое телевидение (Streaming TV) и каталог фильмов (Streaming Movies).

За услуги клиенты могут платить ежемесячно или раз в 1–2 года. 

Доступны различные способы расчёта и возможность получить электронный чек.

#### Создадим новый признак  "ContractDuration"

```python
# Преобразуем даты в to_datetime
contract_df["BeginDate"] = pd.to_datetime(contract_df["BeginDate"], errors="coerce")
contract_df["EndDate"] = pd.to_datetime(contract_df["EndDate"], errors="coerce")

# Вычисляем длительность контракта в днях
contract_df["ContractDuration"] = (contract_df["EndDate"] - contract_df["BeginDate"]).dt.days
```

```python
# Категоризуем длительность контракта 
def categorize_duration(days):
    if days < 180:
        return "Short-term (0-6 months)"
    elif days < 365:
        return "Medium-term (6-12 months)"
    elif days < 730:
        return "Long-term (1-2 years)"
    else:
        return "Very long-term (2+ years)"

contract_df["ContractCategory"] = contract_df["ContractDuration"].apply(categorize_duration)
```

#### Создадим новый категориальный признак "ContractCategory" с классами длительности договора : 0-6 месяцев (короткий); 6-12 месяцев (средний), 1-2 года (длительный) и больше 2 лет (очень длинный)

```python
# Определяем порядок категорий
sorted_categories = ["Short-term (0-6 months)", 
                     "Medium-term (6-12 months)", 
                     "Long-term (1-2 years)", 
                     "Very long-term (2+ years)"]
# Указываем порядок категорий в столбце Contract_Category
contract_df["ContractCategory"] = pd.Categorical(
                                    contract_df["ContractCategory"], 
                                    categories=sorted_categories, 
                                    ordered=True)
# Устанавливаем размер фигуры
plt.figure(figsize=(8, 5))

# Создаем бар-диаграмму для категорий контрактов
sns.countplot(
        data=contract_df, 
        x="ContractCategory", 
        hue="ContractCategory", 
        palette="viridis",
        order=sorted_categories,
        legend=False)

# Настройки заголовка и подписей
plt.title("Распределение длительности контрактов", fontsize=14, fontweight='bold')
plt.xlabel("Категория контракта", fontsize=12)
plt.ylabel("Количество контрактов", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)

# Отображаем график
plt.show()
```

Более половины контрактов длятся более 2-х лет.

## Подготовка данных для обучения

```python
# Определим конечный датафрейм, который будем использовать для обучения и тестирования
final_df = merged_df.drop(columns=[
                                    "CustomerID", 
                                    "EndDate", 
                                    "ContractCategory",
                                    "BeginDate"
                                    ])
final_df.shape
```

```python
# Создадим тренировочную и тестовую выборки
X = final_df.drop("EndDateCntr", axis = 1)
y = final_df["EndDateCntr"]

# Делим на тренировочную и тестовую выборки 
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=RANDOM_STATE, 
                                                    stratify=y)

# Проверяем размеры полученных выборок
print(f"Объем тренировочной выборки: {X_train.shape}")
print(f"Объем тестовой выборки: {X_test.shape}")
```

```python
# Делим тренировочную выборку на окончательную тренировочную (80%) и валидационную (20%)
X_sub_train, X_val, y_sub_train, y_val = train_test_split(X_train, 
                                                  y_train, 
                                                  test_size=0.2, 
                                                  random_state=RANDOM_STATE, 
                                                  stratify=y_train)

# Проверяем размеры выборок
print(f"Объем окончательной тренировочной выборки: {X_sub_train.shape}")
print(f"Объем валидационной выборки: {X_val.shape}")
```

### Подготовка числовых и категориальных признаков

```python
# Выделяем числовые и категориальные признаки
num_features = ["MonthlyCharges", "TotalCharges", "ContractDuration"]
cat_features = ["Type", "PaperlessBilling", "PaymentMethod", "Partner", "Gender", "Dependents", 
                "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", 
                "TechSupport", "StreamingTV", "StreamingMovies"]

# Создаем пайплайн предобработки
num_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),  # Заполняем NaN медианой
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),  # Заполняем NaN "missing"
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", num_transformer, num_features),
    ("cat", cat_transformer, cat_features)
])
```

### Выбор модели, подбор гиперпараметров 

#### Для обучения выберем модели: RandomForestClassifier, XGBClassifier и нейронную сеть.
Поскольку выборка не сбалансирована по целевому признаку, используем SMOTE для RandomForestClassifier, XGBClassifier и подбор параметра scale_pos_weight для XGBClassifier

```python
# Определяем уровни SMOTE для тестирования
smote_ratios = [0.2, 0.4, 0.6, 0.8]

# Поиск лучшей модели для SMOTE
best_smote_model = None
best_smote_classifier_name = ""
best_smote_hyperparams = {}
best_smote_roc_auc = 0
best_smote_accuracy = 0

print("\n Тестируем модели с SMOTE")
for smote_ratio in smote_ratios:
    print(f"\n SMOTE={smote_ratio}")
    
    X_train_transformed = preprocessor.fit_transform(X_sub_train)
    smote = SMOTE(sampling_strategy=smote_ratio, random_state=RANDOM_STATE)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_transformed, y_sub_train)
    
    rf_pipeline = RandomForestClassifier(random_state=RANDOM_STATE, class_weight="balanced")
    rf_params = {
        'n_estimators': [100, 200, 300],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5]
    }
    rf_grid = GridSearchCV(rf_pipeline, rf_params, cv=5, scoring='roc_auc', n_jobs=-1)
    rf_grid.fit(X_train_resampled, y_train_resampled)
    y_pred_rf_proba = rf_grid.best_estimator_.predict_proba(preprocessor.transform(X_val))[:, 1]
    roc_auc_rf = roc_auc_score(y_val, y_pred_rf_proba)
    accuracy_rf = accuracy_score(y_val, (y_pred_rf_proba >= 0.5).astype(int))
    print(f"RandomForest - ROC AUC: {roc_auc_rf:.4f} | Accuracy: {accuracy_rf:.4f}")
    
    xgb_pipeline = XGBClassifier(
            eval_metric="logloss", 
            max_delta_step=1, 
            scale_pos_weight=1, 
            random_state=RANDOM_STATE
        )
    
    xgb_params = {
        'n_estimators': [150, 250],
        'max_depth': [1, 3, 5],
        'learning_rate': [0.05, 0.1],
        'subsample': [0.8, 1.0],
        'colsample_bytree': [0.8, 1.0],
        'gamma': [0.2, 0.3],
        'reg_lambda': [1, 10]
    }
    xgb_grid = GridSearchCV(xgb_pipeline, xgb_params, cv=5, scoring='roc_auc', n_jobs=-1)
    xgb_grid.fit(X_train_resampled, y_train_resampled)
    y_pred_xgb_proba = xgb_grid.best_estimator_.predict_proba(preprocessor.transform(X_val))[:, 1]
    roc_auc_xgb = roc_auc_score(y_val, y_pred_xgb_proba)
    accuracy_xgb = accuracy_score(y_val, (y_pred_xgb_proba >= 0.5).astype(int))
    print(f"XGBoost - ROC AUC: {roc_auc_xgb:.4f} | Accuracy: {accuracy_xgb:.4f}")
    
    if max(roc_auc_rf, roc_auc_xgb) > best_smote_roc_auc:
        best_smote_roc_auc = max(roc_auc_rf, roc_auc_xgb)
        best_smote_accuracy = max(accuracy_rf, accuracy_xgb)
        best_smote_model = rf_grid.best_estimator_ if roc_auc_rf > roc_auc_xgb else xgb_grid.best_estimator_
        best_smote_hyperparams = rf_grid.best_params_ if roc_auc_rf > roc_auc_xgb else xgb_grid.best_params_
        best_smote_classifier_name = "Random Forest" if roc_auc_rf > roc_auc_xgb else "XGBoost"
        # Сохраняем лучшее значение SMOTE-ratio
        best_smote_ratio = smote_ratio

# Выводим результаты лучшей модели с SMOTE
print("\n Лучшая модель с SMOTE:")
print(f" Модель: {best_smote_classifier_name}")
print(f" Лучший ROC AUC: {best_smote_roc_auc:.4f}")
print(f" Лучшая Accuracy: {best_smote_accuracy:.4f}")
print("\n Лучшие гиперпараметры:")
for param, value in best_smote_hyperparams.items():
    print(f"   {param}: {value}")
```

```python
# Определяем уровни weights для тестирования
scale_pos_weights = [1.5, 2, 2.5, 3, 4, 6, 8, 10]

# Поиск лучшей модели для scale_pos_weight
best_xgb_scale_model = None
best_xgb_scale_hyperparams = {}
best_xgb_scale_roc_auc = 0
best_xgb_scale_accuracy = 0

print("\n Тестируем XGBoost с scale_pos_weight")
for weight in scale_pos_weights:
    print(f"\n scale_pos_weight={weight}")
    
    X_train_transformed = preprocessor.fit_transform(X_sub_train)
    xgb_pipeline = XGBClassifier(
            eval_metric="logloss", 
            max_delta_step=1, 
            scale_pos_weight=weight, 
            random_state=RANDOM_STATE
        )
    
    xgb_grid = GridSearchCV(xgb_pipeline, xgb_params, cv=5, scoring='roc_auc', n_jobs=-1)
    xgb_grid.fit(X_train_transformed, y_sub_train)
    y_pred_xgb_proba = xgb_grid.best_estimator_.predict_proba(preprocessor.transform(X_val))[:, 1]
    roc_auc_xgb = roc_auc_score(y_val, y_pred_xgb_proba)
    accuracy_xgb = accuracy_score(y_val, (y_pred_xgb_proba >= 0.5).astype(int))
    print(f"XGBoost - ROC AUC: {roc_auc_xgb:.4f} | Accuracy: {accuracy_xgb:.4f}")
    
    if roc_auc_xgb > best_xgb_scale_roc_auc:
        best_xgb_scale_roc_auc = roc_auc_xgb
        best_xgb_scale_accuracy = accuracy_xgb
        best_xgb_scale_model = xgb_grid.best_estimator_
        best_xgb_scale_hyperparams = xgb_grid.best_params_
        #  Сохраняем лучшее значение scale_pos_weight
        best_xgb_scale_weight = weight

# Выводим результаты лучшей модели с scale_pos_weight
print("\n Лучшая модель XGBoost с scale_pos_weight:")
print(f" Лучший ROC AUC: {best_xgb_scale_roc_auc:.4f}")
print(f" Лучшая Accuracy: {best_xgb_scale_accuracy:.4f}")
print("\n Лучшие гиперпараметры:")
for param, value in best_xgb_scale_hyperparams.items():
    print(f"   {param}: {value}")
```

```python
# Устройство
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Преобразование данных через preprocessor
X_sub_train_transformed = preprocessor.fit_transform(X_sub_train)
X_val_transformed = preprocessor.transform(X_val)

# Преобразование данных в PyTorch тензоры
X_sub_train_tensor = torch.FloatTensor(X_sub_train_transformed)
X_val_tensor = torch.FloatTensor(X_val_transformed)
y_sub_train_tensor = torch.FloatTensor(y_sub_train.values).view(-1, 1)
y_val_tensor = torch.FloatTensor(y_val.values).view(-1, 1)

print(f'Форма тренировочной выборки {X_sub_train_tensor.shape}')
print(f'Форма валидационной выборки {X_sub_train_tensor.shape}')
```

```python
# Определение моделей
def define_model_ReLU(input_dim, n_hidden_1, n_hidden_2, dropout_rate):
    return nn.Sequential(
        nn.Linear(input_dim, n_hidden_1),
        nn.ReLU(),
        nn.Linear(n_hidden_1, n_hidden_2),
        nn.ReLU(),
        nn.Dropout(dropout_rate),
        nn.Linear(n_hidden_2, 1),
        nn.Sigmoid()
    )

def define_model_LeakyReLU(input_dim, n_hidden_1, n_hidden_2, dropout_rate):
    return nn.Sequential(
        nn.Linear(input_dim, n_hidden_1),
        nn.LeakyReLU(),
        nn.Linear(n_hidden_1, n_hidden_2),
        nn.LeakyReLU(),
        nn.Dropout(dropout_rate),
        nn.Linear(n_hidden_2, 1),
        nn.Sigmoid()
    )

def define_model_ReLU_BatchNorm1d(input_dim, n_hidden_1, n_hidden_2, dropout_rate):
    return nn.Sequential(
        nn.Linear(input_dim, n_hidden_1),
        nn.BatchNorm1d(n_hidden_1),
        nn.ReLU(),
        nn.Linear(n_hidden_1, n_hidden_2),
        nn.BatchNorm1d(n_hidden_2),
        nn.ReLU(),
        nn.Dropout(dropout_rate),
        nn.Linear(n_hidden_2, 1),
        nn.Sigmoid()
    )

def define_model_LeakyReLU_BatchNorm1d(input_dim, n_hidden_1, n_hidden_2, dropout_rate):
    return nn.Sequential(
        nn.Linear(input_dim, n_hidden_1),
        nn.BatchNorm1d(n_hidden_1),
        nn.LeakyReLU(),
        nn.Linear(n_hidden_1, n_hidden_2),
        nn.BatchNorm1d(n_hidden_2),
        nn.LeakyReLU(),
        nn.Dropout(dropout_rate),
        nn.Linear(n_hidden_2, 1),
        nn.Sigmoid()
    )
    
# Словарь моделей 
model_variants = {
    "ReLU": define_model_ReLU,
    "LeakyReLU": define_model_LeakyReLU,
    "ReLU_BatchNorm1d": define_model_ReLU_BatchNorm1d,
    "LeakyReLU_BatchNorm1d": define_model_LeakyReLU_BatchNorm1d,
}
```

```python
# Функция обучения и оценки
def train_and_evaluate(
        model_func, 
        n_hidden_1, 
        n_hidden_2, 
        dropout_rate, 
        batch_size, 
        epochs=100, 
        lr=0.005, 
        patience=10):
    
    model = model_func(X_sub_train_tensor.shape[1], n_hidden_1, n_hidden_2, dropout_rate).to(device)
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    criterion = nn.BCELoss()

    dataset = TensorDataset(X_sub_train_tensor, y_sub_train_tensor)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=True, drop_last=True)

    best_roc_auc, best_accuracy = 0, 0
    no_improve_epochs = 0
    best_model_state = None  

    for epoch in range(epochs):
        model.train()
        for batch_X, batch_y in loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            preds = model(batch_X).squeeze(1)
            loss = criterion(preds, batch_y.squeeze(1))

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Оценка каждые 10 эпох
        if (epoch + 1) % 10 == 0:
            model.eval()
            with torch.no_grad():
                val_preds_proba = model(X_val_tensor.to(device)).cpu().numpy().flatten()
                val_preds = (val_preds_proba >= 0.5).astype(int)

                roc_auc = roc_auc_score(y_val_tensor.cpu().numpy(), val_preds_proba)
                accuracy = accuracy_score(y_val_tensor.cpu().numpy(), val_preds)

                print(f'Epoch [{epoch+1}/{epochs}] | '
                      f'Batch: {batch_size} | ROC AUC: {roc_auc:.4f} | Accuracy: {accuracy:.4f}')

                if roc_auc > best_roc_auc:
                    best_roc_auc, best_accuracy = roc_auc, accuracy
                    best_model_state = model.state_dict()
                    no_improve_epochs = 0  
                else:
                    no_improve_epochs += 1

                if no_improve_epochs >= patience:
                    print(f"  Остановка на {epoch+1} эпохе (ранняя остановка)")
                    break

    if best_model_state:
        model.load_state_dict(best_model_state)

    return best_roc_auc, best_accuracy
```

```python
# Функция поиска гиперпараметров
def run_hyperparameter_search(model_name, hyperparams):
    results = []
    best_overall_roc_auc = 0
    best_overall_accuracy = 0
    best_hidden_1, best_hidden_2, best_dropout, best_batch_size = None, None, None, None

    # Выбираем соответствующую модель
    model_func = model_variants[model_name]

    for (n_hidden_1, n_hidden_2), dropout_rate, batch_size in product(
        hyperparams["hidden_layer_sizes"], hyperparams["dropout_rates"], hyperparams["batch_sizes"]
    ):
        print(f'\n Тренировка: Model = {model_name}, Hidden Layers = {n_hidden_1}, {n_hidden_2}, '
              f'Dropout = {dropout_rate}, Batch Size = {batch_size}')
        
        val_roc_auc, val_accuracy = train_and_evaluate(
            model_func, n_hidden_1, n_hidden_2, dropout_rate, batch_size
        )

        # Проверяем, является ли это лучший результат
        if val_roc_auc > best_overall_roc_auc:
            best_overall_roc_auc, best_overall_accuracy = val_roc_auc, val_accuracy
            best_hidden_1, best_hidden_2 = n_hidden_1, n_hidden_2
            best_dropout, best_batch_size = dropout_rate, batch_size

        results.append({
            'Model': model_name,
            'H_Layer1': n_hidden_1,
            'H_Layer2': n_hidden_2,
            'Dropout': dropout_rate,
            'Batch': batch_size,
            'ROC AUC': val_roc_auc,
            'Accuracy': val_accuracy
        })

    # Вывод лучших параметров
    print(f"\n Лучшие параметры: Model = {model_name}, Hidden Layer 1 = {best_hidden_1}, "
          f"Hidden Layer 2 = {best_hidden_2}, Dropout = {best_dropout}, Batch Size = {best_batch_size}, "
          f"ROC AUC: {best_overall_roc_auc:.4f}, Accuracy: {best_overall_accuracy:.4f}")

    # Преобразуем список в DataFrame
    results_df = pd.DataFrame(results)

    # Ограничиваем число знаков после запятой для удобства
    results_df["ROC AUC"] = results_df["ROC AUC"].round(4)
    results_df["Accuracy"] = results_df["Accuracy"].round(4)

    # Вывод всех результатов
    print("\n Результаты перебора параметров:")
    print(results_df)

    return results_df, {
        'Model': model_name,
        'Hid Layer1': best_hidden_1,
        'Hid Layer2': best_hidden_2,
        'Dropout': best_dropout,
        'Batch': best_batch_size,
        'ROC AUC': best_overall_roc_auc,
        'Accuracy': best_overall_accuracy
    }
```

```python
# Применяем препроцессинг к тестовой выборке
X_test_transformed = preprocessor.transform(X_test)

# Делаем предсказания вероятностей на тестовой выборке
y_pred_test_proba = final_model.predict_proba(X_test_transformed)[:, 1]

# Преобразуем вероятности в бинарные предсказания (порог 0.5)
y_pred_test = (y_pred_test_proba >= 0.5).astype(int)

# Оцениваем результаты
roc_auc_test = roc_auc_score(y_test, y_pred_test_proba)
accuracy_test = accuracy_score(y_test, y_pred_test)

# Выводим результаты
print("\n Оценка финальной модели XGBoost на тестовой выборке:")
print(f" ROC AUC: {roc_auc_test:.4f}")
print(f" Accuracy: {accuracy_test:.4f}")
```

```python
# Вычисляем Precision, Recall и пороги
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_test_proba)

# Строим график
plt.figure(figsize=(8, 4))
plt.plot(recall, precision, marker='.', label="Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid()
plt.show()
```

График плавно падает вниз, однако в зоне Recall (0.4 - 0.6), где часто важно поддерживать баланс между Precision и Recall он снижается плавнее.

```python
plt.figure(figsize=(8, 4))
fpr, tpr, _ = roc_curve(y_test, y_pred_test_proba)
plt.plot(fpr, tpr, label="ROC curve", linewidth=2)
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")  # Линия случайных предсказаний
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid()
plt.show()
```

График демонстрирует хорошее качество модели, так как сильно отклоняется от диагональной линии случайного угадывания.

#### Вывод:
В случае компании Теледом Recall (Полнота) важнее, поскольку для него важно заранее определить пользователей, которые, скорее всего, разорвут договор.
	•	Если модель говорит, что пользователь уйдет, а он не уходит (ложноположительный), это не так критично. Можно предложить ему скидку или бонус.
	•	Но если модель пропустит пользователя, который действительно разорвет договор (ложноотрицательный), компания потеряет клиента.

Поэтому можно изменить порог для более полного охвата клиентов, которые могут не продлить договор. Точность (Precision) при этом может снизиться, но это допустимо, если приоритет — не упустить “уходящих” клиентов.

```python
# Создаем список порогов от 0.1 до 0.9 (с шагом 0.05)
thresholds = np.arange(0.1, 0.9, 0.05)

best_threshold = 0.5  # Начальный вариант (стандартный порог)
best_f1 = 0  # Лучший F1-score

# Перебираем пороги
for threshold in thresholds:
    y_pred_test = (y_pred_test_proba >= threshold).astype(int)
    recall = recall_score(y_test, y_pred_test)
    precision = precision_score(y_test, y_pred_test)
    f1 = f1_score(y_test, y_pred_test)

    print(f"Порог: {threshold:.2f} | Recall: {recall:.4f} | Precision: {precision:.4f} | F1-score: {f1:.4f}")

    # Сохраняем лучший порог по F1-score
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold

# Выводим лучший порог
print(f"\n Оптимальный порог: {best_threshold:.2f} с F1-score = {best_f1:.4f}")
```

#### Оценим значения ROC AUC и Accuracy для оптимального порога

```python
# Применяем оптимальный порог
y_pred_optimal = (y_pred_test_proba >= best_threshold).astype(int)

# Вычисляем метрики
roc_auc_optimal = roc_auc_score(y_test, y_pred_test_proba) 
accuracy_optimal = accuracy_score(y_test, y_pred_optimal)

print(f" ROC AUC для оптимального порога: {roc_auc_optimal:.4f}")
print(f" Accuracy для оптимального порога: {accuracy_optimal:.4f}")
```

#### Анализ, какой порог лучше выбрать для охвата максимального количества уходящих клиентов?
    •	Максимальный Recall = 0.9273 при пороге 0.10, но Precision очень низкая (0.3446) – будет много ложных срабатываний.
	•	Recall = 0.8691 при пороге 0.15 – Precision немного выше (0.4230).
	•	Recall = 0.8109 при пороге 0.20 – еще лучшее соотношение с Precision (0.4945).

Можно взять порог 0.15, потому что:
1.	Recall 86.9% – достаточно высокий охват.
2.	Precision 42.3% – приемлемая точность (не слишком много ложных тревог).
3.	F1-score 56.9% – хороший баланс между Precision и Recall.

#### Оценим значения ROC AUC и Accuracy для порога 0.15

```python
threshold = 0.15
y_pred_test = (y_pred_test_proba >= threshold).astype(int)
roc_auc = roc_auc_score(y_test, y_pred_test_proba)
print(f"ROC AUC при пороге {threshold}: {roc_auc:.4f}")
accuracy = accuracy_score(y_test, y_pred_test)
print(f"Accuracy при пороге {threshold}: {accuracy:.4f}")
```
